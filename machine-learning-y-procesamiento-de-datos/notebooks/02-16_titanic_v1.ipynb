```python
import pandas as pd
import numpy as np
```

```python
# Cargar datos
df = pd.read_csv("../datasets/titanic_train.csv")
```

```python
# Previsualización de la base de datos (primeros 5 datos)
df.head()
```

```python
# Cantidad de datos faltantes por columna
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer_mean = SimpleImputer(strategy="mean")

df_mean = df.copy()
df_mean[["Age"]] = imputer_mean.fit_transform(df_mean[["Age"]]) # Los doble corchetes es para referencial la columna, solo lo hace pandas

df_mean[["Age"]].isnull().sum()

print(df_mean[["Age"]])
```

```python
imputer_median = SimpleImputer(strategy="median")

df_median = df.copy()
df_median[["Age"]] = imputer_median.fit_transform(df_median[["Age"]]) # Los doble corchetes es para referencial la columna, solo lo hace pandas

df_median[["Age"]].isnull().sum()

print(df_median[["Age"]])
```

```python
from sklearn.impute import KNNImputer

imputer_knn = KNNImputer(n_neighbors=5)
df_knn = imputer_knn.fit_transform(df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']])

print("Valor original (fila 5):", df.iloc[5]['Age']) # Suele ser NULO o NaN
#print("Imputación Media:", df_median["Age"])
#print("Imputación Promedio:", df_mean["Age"])
print("Imputación KNN:", df_knn[5,2])
```

```python
from category_encoders import TargetEncoder, HashingEncoder
```

```python
df_cat = pd.read_csv("../datasets/titanic_train.csv")[['Embarked', 'Cabin', 'Survived']]
df_cat['Cabin'] = df_cat['Cabin'].fillna('Unknown')

# Target Encoding para Embarcado
te = TargetEncoder(cols=['Embarked'])
df_target = te.fit_transform(df_cat['Embarked'], df_cat['Survived'])

# Hashing Encoding para camarote
he = HashingEncoder(cols=['Cabin'], n_components=5)
df_hash = he.fit_transform(df_cat['Cabin'])

print(df_target.head())
print(df_hash.head())
```
