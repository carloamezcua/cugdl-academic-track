```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Cargar datos
df = pd.read_csv("datasets/estudio_vs_calificacione.csv")
```

```python
# Variables
X = df[['horas_estudio']]
y = df['calificacion']
```

```python
# Crear y entrenar modelo
model = LinearRegression()
model.fit(X, y)
```

```python
# Parámetros del modelo
m = model.coef_[0]
b = model.intercept_

print(f"Pendiente (m): {m}")
print(f"Intercepto (b): {b}")
```

```python
# Predicción
horas = 7
prediccion = model.predict([[horas]])
print(f"Calificación estimada para {horas} horas: {prediccion[0]:.2f}.")
```

```python
# Visualización
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("Horas de estudio")
plt.ylabel("Calificación")
plt.title("Regresión lineal: Horas de estudio vs Calificación")
plt.show()
```
